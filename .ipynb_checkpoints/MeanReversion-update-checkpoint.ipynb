```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from arch import arch_model
from datetime import timedelta
import seaborn as sns
import mplfinance as mpf
import yfinance as yf


# Descargar datos de Apple (AAPL)

ticker = 'EURUSD=X'
df = yf.download(ticker, period='5d', interval='5m')
```

```python
# Visualizar la serie de precios de cierre

plt.figure(figsize=(12, 5))
plt.plot(df['Close'], label='Precio de Cierre')
plt.title(f'Serie de precios de {ticker}')
plt.xlabel('Fecha')
plt.ylabel('Precio ($)')
plt.legend()
plt.grid(True)
plt.show()

# --- CALCULAR RETORNOS LOGARTMICOS ---
df['LogReturn'] = np.log(df['Close'] / df['Close'].shift(1))

# --- CALCULAR MEDIA Y DESVIACIN ESTNDAR MVIL ---
window = 20
df['mean'] = df['LogReturn'].rolling(window).mean()
df['std'] = df['LogReturn'].rolling(window).std()

# --- BANDAS DE +2 DESVIACIONES ---
df['upper'] = df['mean'] + 2 * df['std']
df['lower'] = df['mean'] - 2 * df['std']


# Visualizar los rendimientos logartmicos

plt.figure(figsize=(12, 5))
plt.plot(df['LogReturn'], label='Rendimiento Logartmico', color='orange')
plt.title(f'Rendimientos Logartmicos Diarios de {ticker}')
plt.xlabel('Fecha')
plt.ylabel('Log Return')
plt.legend()
plt.grid(True)
plt.show()

# --- Confirmacin por 2 velas consecutivas ---
df['above_upper'] = df['LogReturn'] > df['upper']
df['below_lower'] = df['LogReturn'] < df['lower']

# --- Umbral mnimo de magnitud ---
threshold = 0.0008
df['evento'] = pd.Series(index=df.index, dtype="object")
df.loc[(df['above_upper'].rolling(2).sum() == 2) &
       (df['LogReturn'] > threshold), 'evento'] = 'Reversin (+2)'
df.loc[(df['below_lower'].rolling(2).sum() == 2) &
       (df['LogReturn'] < -threshold), 'evento'] = 'Reversin (-2)'

# --- GRAFICAR ---
plt.figure(figsize=(14, 5))
plt.plot(df['LogReturn'], label='Retorno logartmico suavizado (m5)', color='orange')
plt.plot(df['upper'], linestyle='--', color='red', label='+2')
plt.plot(df['lower'], linestyle='--', color='green', label='-2')
plt.axhline(0, linestyle='--', color='gray', linewidth=1)

# Marcar puntos de evento
eventos = df.dropna(subset=['evento'])
plt.scatter(eventos.index, eventos['LogReturn'], color='purple', marker='o', s=50)

plt.title("Retornos Logartmicos Suavizados - EURUSD (m5) con Seales Filtradas de 2")
plt.xlabel("Fecha")
plt.ylabel("Log Return")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
