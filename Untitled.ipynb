Transformacion a log-retornos

```python
# Calcula los log-rendimientos
df["log_return"] = np.log(df["Close"] / df["Close"].shift(1))

# --- LIMPIEZA CRÍTICA DE log_return ---
# Reemplaza explícitamente los infinitos con NaN, luego elimina todos los NaN
log_return_cleaned = df["log_return"].replace([np.inf, -np.inf], np.nan).dropna()

# ****** VERIFICACIÓN FINAL DE LA SERIE LIMPIA ******
print("\n--- Verificación de log_return antes de ARIMA ---")
print(f"Número de NaNs en la serie limpia: {log_return_cleaned.isna().sum()}")
print(f"¿Hay infinitos en la serie limpia?: {np.isinf(log_return_cleaned).any()}")
print(f"Número de observaciones en la serie limpia: {len(log_return_cleaned)}")
print(f"Varianza de la serie limpia: {log_return_cleaned.var()}")
print("Primeros 5 valores de la serie limpia:\n", log_return_cleaned.head())
print("Últimos 5 valores de la serie limpia:\n", log_return_cleaned.tail())

if log_return_cleaned.empty:
    print("¡ERROR FATAL!: La serie log_return_cleaned está VACÍA después de la limpieza. Revisa tus datos originales.")
elif log_return_cleaned.var() == 0:
    print("¡ERROR FATAL!: La serie log_return_cleaned tiene VARIANZA CERO (todos los valores son iguales). El modelo ARIMA no puede ajustarse.")
```

Prueba Estacionaria - ADF Test

```python
# Muestra los primeros valores de log_return para ver si hay NaNs o Inf.
print(df["log_return"].head())

# Cuenta los NaNs
print(df["log_return"].isna().sum())

# Verifica si hay infinitos (True si hay al menos un infinito)
print(np.isinf(df["log_return"]).any())

# Muestra las filas con infinitos (si las hay)
print(df[np.isinf(df["log_return"])])

result = adfuller(df["log_return"].dropna())
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
```

p-value = 0.00000000000000000005: Mucho menor que cualquier nivel típico de significancia (0.05, 0.01, etc.).

✅ Conclusión: La serie de log-retornos es estacionaria, por lo tanto:

No es necesario diferenciar.

Siguiente paso: ACF y PACF para identificar estructura ARIMA

```python
print("Number of NaNs in log_return before plotting:", df['log_return'].isna().sum())
print("Are there any infinite values in log_return before plotting?", np.isinf(df['log_return']).any())
print("First 5 values of log_return (after any previous cleaning):")
print(df['log_return'].head())
print("Last 5 values of log_return (after any previous cleaning):")
print(df['log_return'].tail())

# Re-calculate log_return (just in case it was overwritten or corrupted)
df["log_return"] = np.log(df["Close"] / df["Close"].shift(1))

# --- Cleaning step for log_return ---
# Convert inf/-inf to NaN, then drop all NaNs
log_return_cleaned = df["log_return"].replace([np.inf, -np.inf], np.nan).dropna()

# --- Verification of cleaned series ---
print("\n--- After cleaning for ACF/PACF ---")
print("Number of NaNs in cleaned log_return:", log_return_cleaned.isna().sum())
print("Are there any infinite values in cleaned log_return?", np.isinf(log_return_cleaned).any())
print("Shape of cleaned log_return:", log_return_cleaned.shape)
if log_return_cleaned.empty:
    print("WARNING: log_return_cleaned is empty after cleaning. Cannot plot ACF/PACF.")
else:
    print("First 5 values of cleaned log_return:")
    print(log_return_cleaned.head())


# --- Plotting using the CLEANED series ---
plot_acf(log_return_cleaned, lags=30)
plt.title("ACF - Log Rendimientos")
plt.show()

plot_pacf(log_return_cleaned, lags=30)
plt.title("PACF - Log-Rendimientos")
plt.show()
```

```python
# Vuelve a crear log_return desde cero por si acaso
df["log_return"] = np.log(df["Close"] / df["Close"].shift(1))

# Limpia la serie de NaNs y infinitos
log_return_para_arima = df["log_return"].replace([np.inf, -np.inf], np.nan).dropna()

# ****** VERIFICACIÓN CRÍTICA ******
print("\n--- Verificación de log_return antes de ARIMA ---")
print("Número de NaNs:", log_return_para_arima.isna().sum())
print("¿Hay infinitos?:", np.isinf(log_return_para_arima).any())
print("Primeros 5 valores:", log_return_para_arima.head())
print("Últimos 5 valores:", log_return_para_arima.tail())
print("Estadísticas descriptivas:\n", log_return_para_arima.describe())
print("¿Todos los valores son cero o casi cero?:", (log_return_para_arima.abs() < 1e-9).all())
print("Variancia:", log_return_para_arima.var())

if log_return_para_arima.empty:
    print("ERROR: La serie log_return_para_arima está vacía después de la limpieza.")
    # Revisa el archivo original o la creación de log_return
elif log_return_para_arima.var() == 0:
    print("ERROR: La serie log_return_para_arima tiene varianza cero (todos los valores son iguales). No se puede ajustar ARIMA.")
    # Esto podría pasar si todos los retornos son idénticos o cero.

model_arima = ARIMA(df['log_return'].dropna(), order=(1, 0, 1))
result_arima = model_arima.fit()
print(result_arima.summary())
```

Coeficientes del modelo: AR(1) = -0.6105, MA(1) = -0.4681 -> Ambos son significativos (p < 0.05), así que el modelo tiene algo de estructura.

const = 0.0004 no es significativo, pero tampoco es problema.

✅ Diagnósticos: Ljung-Box (Q) p-value = 0.81 -> Los residuos no tienen autocorrelación, ✅ buen síntoma.

Jarque-Bera (JB) p-value = 0.00 -> Los residuos no son normales, lo cual es típico en retornos financieros.

Heteroskedasticity (ARCH test) p-value = 0.00 -> Hay heterocedasticidad en los residuos.

⚠️ Conclusión: Aunque el ARIMA(1,0,1) ajusta razonablemente la media:

✅ No hay autocorrelación en los residuos.

❌ Hay heterocedasticidad -> necesitamos GARCH.

```python
garch_model = arch_model(df['log_return'].dropna(), vol='GARCH', p=1, q=1, mean='ARX', lags=1)
garch_result = garch_model.fit(disp='off')
print(garch_result.summary())
```

✅ Diagnóstico Rápido:
Check Resultado ¿Coeficientes significativos? ✅ Todos ¿Persistencia (α+β < 1)? ✅ Sí, cerca de 1 pero estable ¿Modelo razonable para riesgo/volatilidad? ✅ Perfect

```python
# Forecast a 1 año (252 días hábiles)

# # Forecast 252 días hacia adelante (1 año de trading)
forecast_horizon = 252
forecast = garch_result.forecast(horizon=forecast_horizon)

# # Extraemos las trayectorias completas (última fila: el forecast futuro desde hoy)
mean_forecast = forecast.mean.iloc[-1].values
variance_forecast = forecast.variance.iloc[-1].values
vol_forecast = np.sqrt(variance_forecast)

# # Impresiones correctas
print(f"Primera media forecast: {mean_forecast[0]:.6f}")
print(f"Última media forecast: {mean_forecast[-1]:.6f}")
print(f"Primera desviación estándar forecast: {vol_forecast[0]:.6f}")
print(f"Última desviación estándar forecast: {vol_forecast[-1]:.6f}")
```

```python
# Suponer que tenés df['SP500'] como serie histórica de precios
last_price = df['Close'].iloc[-1]

# 1. Cálculo acumulado
cum_mean = np.cumsum(mean_forecast)
cum_std = np.sqrt(np.cumsum(vol_forecast**2))

# 2. Forecast de precios y bandas
expected_price = last_price * np.exp(cum_mean)
upper_band = last_price * np.exp(cum_mean + cum_std)
lower_band = last_price * np.exp(cum_mean - cum_std)

# 3. Fechas futuras
future_dates = pd.date_range(start=df.index[-1], periods=forecast_horizon + 1, freq='B')[1:]

# 4. Gráfico
plt.figure(figsize=(14, 6))

# Historial
plt.plot(df.index, df['Close'], label='Histórico XAUUSD', color='black')

# Forecast
plt.plot(future_dates, expected_price, label='Forecast GARCH', color='blue')

# Bandas de confianza 1de
plt.fill_between(future_dates, lower_band, upper_band, color='skyblue', alpha=0.4, label='±1 Desv. Estándar')

# Detalles
plt.title('Forecast XAUUSD a 1 año con bandas de confianza (GARCH)')
plt.xlabel('Fecha')
plt.ylabel('Precio estimado')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
